```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio

# Ensure Order_date is a datetime column
data['Order_date'] = pd.to_datetime(data['Order_date'])

# Pie Chart for Sum of Sales per Product
sales_per_product = data.groupby('Product')['Sales'].sum().reset_index()
fig_pie = px.pie(sales_per_product, values='Sales', names='Product', title='Sum of Sales per Product')
fig_pie.show()

# Extract month and year from Order_date
data['YearMonth'] = data['Order_date'].dt.to_period('M')

# Line Chart for Count of Orders per Month
orders_per_month = data.groupby('YearMonth')['Document_number'].nunique().reset_index()
orders_per_month['YearMonth'] = orders_per_month['YearMonth'].dt.to_timestamp()
fig_orders = px.line(orders_per_month, x='YearMonth', y='Document_number', title='Count of Orders per Month')
fig_orders.show()

# Line Chart for Count of Clients per Month
clients_per_month = data.groupby('YearMonth')['Partner_code'].nunique().reset_index()
clients_per_month['YearMonth'] = clients_per_month['YearMonth'].dt.to_timestamp()
fig_clients = px.line(clients_per_month, x='YearMonth', y='Partner_code', title='Count of Clients per Month')
fig_clients.show()

# Line Chart for Count of Products per Month
products_per_month = data.groupby('YearMonth')['Product'].nunique().reset_index()
products_per_month['YearMonth'] = products_per_month['YearMonth'].dt.to_timestamp()
fig_products = px.line(products_per_month, x='YearMonth', y='Product', title='Count of Products per Month')
fig_products.show()
```
